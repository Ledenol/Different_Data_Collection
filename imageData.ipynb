```python
!pip install opencv-python pandas --quiet
import cv2
import os
import pandas as pd


folder = r"C:\Users\KIIT0001\Downloads"

image_data = []
image_names = []

for file in os.listdir(folder):
    if file.lower().endswith((".png", ".jpg", ".jpeg")):
        path = os.path.join(folder, file)
        img = cv2.imread(path)
        if img is not None:
            image_data.append(img)
            image_names.append(file)

print("Images loaded:", len(image_data))

pd.DataFrame({
    "filename": image_names,
    "shape (H,W,C)": [img.shape for img in image_data]
})
```
